```python
import cv2 
import numpy as np
```

```python
identity = np.array([[0.,0.,0.],
                     [0.,1.,0.],
                     [0.,0.,0.]])
# Sharpen
sharpen = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0]])
# Gaussian Blur
gaussian = (1 / 16.0) * np.array([[1., 2., 1.],
                                  [2., 4., 2.],
                                  [1., 2., 1.]])

gaussian_5x5 = (1/256.0)*np.array([[1., 4., 6., 4., 1.],
                                   [4., 16., 24., 16., 4.],
                                   [6., 24., 36., 24., 6.],
                                   [4., 16., 24., 16., 4.],
                                   ])
# for edges in X axis
sobel_edge_kernel_x = np.array([[1., 2., 1.],
                              [0., 0., 0.],
                              [-1., -2., -1]])
# for edges in Y axis
sobel_edge_kernel_y = np.array([[1., 0., -1.],
                              [2., 0., -2.],
                              [1., 0., -1.]])
```

```python
image = cv2.imread("frames_sample/original.png")
# image = cv2.imread("frames_sample\original.png")
```

```python
def convoperation(image, kernel, kernel_name):
    result = cv2.filter2D(image, -1, kernel)
    cv2.imshow('Original Image', image)
    cv2.imshow(f'{kernel_name} Image', result)
    cv2.imwrite(f'frames_sample/{kernel_name}_Image.png', result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
convoperation(image,sharpen,"sharpen")
```

```python
convoperation(image,gaussian,"gaussian")
```

```python
convoperation(image,sobel_edge_kernel_x,"sobel_x")
```

```python
convoperation(image,sobel_edge_kernel_y,"sobel_y")
```

```python
def warp(img):
    img_size = (img.shape[1], img.shape[0])

    pt1 = np.float32([[685, 450],
          [1090, 710],
          [220, 710],
          [595, 450]])

    pt2 = np.float32(
        [[900, 0],
          [900, 710],
          [250, 710],
          [250, 0]])

    M = cv2.getPerspectiveTransform(pt1, pt2)
    Minv = cv2.getPerspectiveTransform(pt2, pt1)

    binary_warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)

    return binary_warped, Minv
```

```python
actual_output,_ = warp(actual_output)
height, width, _ = actual_output.shape
actual_output[:, width//2:]=0
predicted_output[:, width//2:]=0
```

```python
lower_red = np.array([0, 0, 240])
upper_red = np.array([255, 255, 255])

mask = cv2.inRange(actual_output, lower_red, upper_red)

actual_red_lane = cv2.bitwise_and(actual_output, actual_output, mask=mask)

actual_gray = cv2.cvtColor(actual_red_lane, cv2.COLOR_BGR2GRAY)
predicted_gray = cv2.cvtColor(predicted_output, cv2.COLOR_BGR2GRAY)

_, actual_binary = cv2.threshold(actual_gray, 10, 255, cv2.THRESH_BINARY)
_, predicted_binary = cv2.threshold(predicted_gray, 200, 255, cv2.THRESH_BINARY)

intersection = np.logical_and(actual_binary, predicted_binary)
union = np.logical_or(actual_binary, predicted_binary)
iou_score = np.sum(intersection) / np.sum(union)

print("Intersection over Union (IoU) score:", iou_score)

intersection_visualization = intersection.astype(np.uint8) * 255
union_visualization = union.astype(np.uint8) * 255

cv2.imwrite('Intersection.png', intersection_visualization)
cv2.imwrite('Union.png', union_visualization)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
diff = cv2.absdiff(actual_binary, predicted_binary)

total_pixels = actual_binary.size
differing_pixels = np.count_nonzero(diff)
percentage_difference = (differing_pixels / total_pixels) * 100

print("Percentage difference:", percentage_difference)
```
